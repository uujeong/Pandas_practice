### 2) 피봇테이블

```python
df.head()
```

```python
df['median_house_value'] < df['median_house_value'].quantile(0.25)
```

```python
df['house_price'] = 'Medium'

df.loc[df['median_house_value'] < df['median_house_value'].quantile(0.25), 'house_price'] = 'Low'
df.loc[df['median_house_value'] > df['median_house_value'].quantile(0.75), 'house_price'] = 'High'

df['house_price'].value_counts()
```

```python
df
```

```python
df_pvt = pd.pivot_table(df,                             # 원본 데이터프레임
                     index='house_price',               # 행 위치
                     columns='ocean_proximity',        # 열 위치
                     values='housing_median_age',      # 데이터
                     aggfunc='mean')                   # 집계 함수

df_pvt
```
